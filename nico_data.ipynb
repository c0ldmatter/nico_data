```python
# Import needed libraries

import pandas as pd
import matplotlib.pyplot as plt


# Data input as table using pandas DataFrame
data = pd.DataFrame ({'Bldg Type': ['C1','C1-0','CHB','CWS','MWS','N','S1','S3','S4','URA','URM','W1','W3','W4','W5'],
                      'Count': [7,4,3631,73,2658,67,4,1,1,1,1,2841,832,305,2250]
                      })
# Define building type with corresponding color
color_map = {
    'C1': '#D7191C',
    'C1-0': '#E13E2D',
    'CHB': '#EA643F',
    'CWS': '#F48950',
    'MWS': '#FEC279',
    'N': '#FED790',
    'S1': '#FFEBA8',
    'S3': '#FFFFBF',
    'S4': '#EAF7B8',
    'URA': '#D5EEB2',
    'URM': '#C0E6AB',
    'W1': '#ABDDA4',
    'W3': '#8BC7AA',
    'W4': '#6BB0AF',
    'W5': '#4B9AB4'
}

# Create a list of colors for each Bldg Type
colors = [color_map[bldg_type] for bldg_type in data['Bldg Type']]

# Plot data using bar graph

plt.figure(figsize=(10,6))

# Add major and minor gridlines
plt.grid(which='major', linestyle='-', linewidth='0.5', color='gray')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='lightgray')

# Set minor ticks on the y-axis
plt.minorticks_on()

plt.bar(data['Bldg Type'], data['Count'], color=colors)
plt.title('Building Type Distribution')
plt.xlabel('Building Type')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
